```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tracks_df = pd.read_csv('tracks.csv')
artists_df = pd.read_csv('artists.csv')
spotify_features_df = pd.read_csv('SpotifyFeatures.csv')
```

```python
most_popular = tracks_df.query('popularity>90',inplace=False).sort_values('popularity',ascending= False)
most_popular
```

```python
print("\nSpotify Features Dataset Info:")
spotify_features_df.info()
print("\nSpotify Features Dataset Description:")
display(spotify_features_df.describe())
```

```python
numeric_df = tracks_df.drop(["key", "mode", "explicit"], axis=1).select_dtypes(include=[float, int])

corr_df = numeric_df.corr(method="pearson")

plt.figure(figsize=(14,6))
heatmap = sns.heatmap(corr_df, annot=True, fmt=".1g", vmin=-1, vmax=1, center=0, cmap="inferno", linewidth=1, linecolor="Black")
heatmap.set_title("Correlation HeatMap Between Variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
plt.show()
```

```python
sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))
sns.histplot(spotify_features_df['popularity'], kde=True)
plt.title('Distribution of Track Popularity')
plt.xlabel('Popularity')
plt.ylabel('Count')
plt.show()

```

```python
sample_df=tracks_df.sample(int(0.004*len(tracks_df)))
```

```python
print(len(sample_df))
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df, y = "loudness", x= "energy", color = "y").set(title ="Loudness VS Energy correlation")
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df, y = "popularity", x= "acousticness", color = "b").set(title ="Popularity VS Acousticness correlation")
```

```python
# Scatter plot of danceability vs. energy
plt.figure(figsize=(10, 6))
sns.scatterplot(x='danceability', y='energy', data=spotify_features_df, hue='popularity', palette='viridis', alpha=0.7)
plt.title('Danceability vs. Energy')
plt.xlabel('Danceability')
plt.ylabel('Energy')
plt.legend(title='Popularity', loc='upper right')
plt.show()
```

```python
# Plot the distribution of genres
plt.figure(figsize=(12, 6))
genres = spotify_features_df['genre'].value_counts().head(20)
sns.barplot(x=genres.index, y=genres.values, palette='viridis')
plt.title('Top 20 Genres')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python

key_stats = spotify_features_df[['danceability', 'energy', 'valence', 'acousticness', 'instrumentalness', 'popularity']].describe()
print("Key Statistics for Tracks:")
display(key_stats)
```

```python
# Plot the average popularity of tracks by genre
plt.figure(figsize=(14, 8))
avg_popularity_by_genre = spotify_features_df.groupby('genre')['popularity'].mean().sort_values(ascending=False).head(20)
sns.barplot(x=avg_popularity_by_genre.index, y=avg_popularity_by_genre.values, palette='viridis')
plt.title('Top 20 Genres by Average Popularity')
plt.xlabel('Genre')
plt.ylabel('Average Popularity')
plt.xticks(rotation=45)
plt.show()
```

```python
# Box plot for distribution of features by genre
features = ['danceability', 'energy', 'instrumentalness']

for feature in features:
    plt.figure(figsize=(14, 8))
    sns.boxplot(x='genre', y=feature, data=spotify_features_df, palette='viridis')
    plt.title(f'Distribution of {feature.capitalize()} by Genre')
    plt.xlabel('Genre')
    plt.ylabel(feature.capitalize())
    plt.xticks(rotation=90)
    plt.show()

```
